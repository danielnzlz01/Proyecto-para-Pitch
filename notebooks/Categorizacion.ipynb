```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Load the dataset
data = df1  # Replace "your_dataset.csv" with your actual dataset file name

# Preprocess the data
X = data['Comment'].tolist()  # Convert to list

# Remove stopwords from comments
stop_words = set(stopwords.words('spanish'))  # Set of stopwords for Spanish, you can use other languages as well
X = pd.Series(X).apply(lambda comment: ' '.join([word for word in word_tokenize(comment) if word.lower() not in stop_words]))

# Feature extraction - convert text comments to numerical features
vectorizer = CountVectorizer()
X_features = vectorizer.fit_transform(X)

# Apply Latent Dirichlet Allocation (LDA)
num_topics = 30  # Number of topics/themes to identify
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X_features)

# Get the most important words for each topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    print(f"Topic #{topic_idx+1}:")
    # Get the indices of the 5 most important words for each topic
    top_words_idx = topic.argsort()[-5:]
    top_words = [feature_names[idx] for idx in top_words_idx]
    print(", ".join(top_words))
    print()

# Predict the main theme for new comments
new_comments = ["Odio la aerolinea, me decepcionan", 
                "Me encanta viajar con ustedes", 
                "Ni me gusta ni me desagrada es una experiencia normal",
                "el servicio de abordaje es lento me canse de esperar", 
                "la tarjeta de credito paso correctamente",
                "estoy satisfecho con mi viaje", 
                "excelente servicio", 
                "la pagina web se traba mucho"]  # Replace with your new comments
new_comments_features = vectorizer.transform(new_comments)
predicted_topics = lda.transform(new_comments_features)
for comment, topics in zip(new_comments, predicted_topics):
    dominant_topic = np.argmax(topics)  # Get the dominant topic for each comment
    print(f"Comment: {comment}")
    print(f"Predicted Topic: {dominant_topic+1}")
    print()
```
